```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Raw Sales Data.csv")
print(df.info())
df.head()
```

```python
df['Sales']=df['Sales'].str.replace(",","")
df['Sales']=df['Sales'].astype(int)
df['Sales'].dtype
```

```python
df['Qty']=df['Qty'].str.replace(",","")
df['Qty']=df['Qty'].astype(int)
df['Qty'].dtype
```

```python
#1. Monthwise number of quantities sold
df_monthwise_Qty=df.groupby(["Month"])["Qty"].sum()
df_monthwise_Qty=pd.DataFrame(df_monthwise_Qty).reset_index(0)
df_monthwise_Qty
```

```python
#2. Category wise Qty sold
df_categorywise_Qty=df.groupby(["Category"])["Qty"].sum()
df_categorywise_Qty=pd.DataFrame(df_categorywise_Qty).reset_index(0)
df_categorywise_Qty
```

```python
#Sub-channel wise total sales
df_subchannelwise_sales=df.groupby(["Sub-Channels"])["Sales"].sum()
df_subchannelwise_sales=pd.DataFrame(df_subchannelwise_sales).reset_index(0)
df_subchannelwise_sales
```

```python
#Highest sale in which month
df_highest=df["Sales"].max()
df_highestsale=df[df["Sales"]==df_highest]

df_highestsale['Month'].values[0]
df_highestsale=df_highestsale['Month']
df_highestsale=pd.DataFrame(df_highestsale.reset_index(0))
df_highestsale
```

```python
#Replacing regex for future version
df['Product Name']= df['Product Name'].str.replace('\d+','#')
df['Product Name']=df['Product Name'].apply(lambda x:x.split('#')[0])
df['Product Name']
```

```python
#5. Highest selling product (based on sales) (packaging quantity doesn't matter 
#Ex. consider "GET SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)

highest_selling=df.groupby('Product Name')['Sales'].sum()
df_highest_selling_product=highest_selling .reset_index().sort_values('Sales', ascending=False).head(1)
df_highest_selling_product
```

```python
df
```

```python
#. List of different products company is offering (packaging quantity doesn't matter. 
#Ex. consider "GET SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)
df_category=df.groupby('Product Name')['Sales'].sum()
df_category=pd.DataFrame(df_category.reset_index(0))
df_category
```

```python
#Summary in excel
writer=pd.ExcelWriter('Raw Sales Data solved by Neeraj Raut.xlsx',engine='xlsxwriter')
df.to_excel(writer,sheet_name='Data',index=False)
df_monthwise_Qty.to_excel(writer,sheet_name='df_monthwise_Qty',index=False)
df_categorywise_Qty.to_excel(writer,sheet_name='df_categorywise_Qty',index=False)
df_subchannelwise_sales.to_excel(writer,sheet_name='df_subchannelwise_sales',index=False)
df_highestsale.to_excel(writer,sheet_name='df_highestsale',index=False)
df_highest_selling_product.to_excel(writer,sheet_name='df_highest_selling_product',index=False)
df_category.to_excel(writer,sheet_name='df_category',index=False)
writer.save()
```
